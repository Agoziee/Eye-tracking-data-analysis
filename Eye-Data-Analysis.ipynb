```python
eye_data = pd.read_csv("eyetracking_data_1.csv")
eye_data.head(5)
```

```python
eye_data.columns
```

```python
# Selecting specific columns needed for analysis
eye_data_imp_col = eye_data[['relative_to_video_first_frame_timestamp', 'gaze_projected_to_left_view_x', 
         'gaze_projected_to_left_view_y', 'gaze_projected_to_right_view_x', 
          'gaze_projected_to_right_view_y']]
eye_data_imp_col.head(5)
```

```python
# Calculating the mean of the left and right eye view on the x & y coordinates respectively
xavg_column = eye_data_imp_col[['gaze_projected_to_left_view_x', 'gaze_projected_to_right_view_x']].mean(axis=1)
yavg_column = eye_data_imp_col[['gaze_projected_to_left_view_y', 'gaze_projected_to_right_view_y']].mean(axis=1)

# Adding the calculated values to the dataframe
eye_data_imp_col.insert(2, "xavg", xavg_column, True)
eye_data_imp_col.insert(6, "yavg", yavg_column, True)

eye_data_imp_col.head(5)
```
