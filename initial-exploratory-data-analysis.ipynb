```python
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
filename = "../data/participant_10.csv"
```

```python
data = pd.read_csv(filename)
```

```python
data
```

```python
data.info()
```

```python
data["datetime"] = data['raw_timestamp'].apply(lambda timestamp: dt.datetime.fromtimestamp(timestamp // 1000000000))
```

```python
data.head()
```

```python
data["datetime"]
```

```python
data.columns.tolist()
```

```python
columns_of_interest = ['datetime',
                       'relative_to_video_first_frame_timestamp', 
                       'gaze_projected_to_left_view_x', 
                       'gaze_projected_to_left_view_y', 
                       'gaze_projected_to_right_view_x', 
                       'gaze_projected_to_right_view_y']
df = data[columns_of_interest]
```

```python
df
```

```python
df.plot(x="datetime", y="gaze_projected_to_left_view_x", figsize=(10,9))
plt.show()
```

```python
df.plot(x="datetime", y="gaze_projected_to_left_view_y", figsize=(10,9))
plt.show()
```

```python
df.plot(x="datetime", y="gaze_projected_to_right_view_x", figsize=(10,9))
plt.show()
```

```python
df.plot(x="datetime", y="gaze_projected_to_right_view_y", figsize=(10,9))
plt.show()
```

```python
import numpy as np

heatmap, xedges, yedges = np.histogram2d(df["gaze_projected_to_left_view_x"], df["gaze_projected_to_left_view_y"], bins=10)
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

plt.clf()
plt.imshow(heatmap.T, extent=extent, origin='lower')
plt.show()
```

```python
heatmap, xedges, yedges = np.histogram2d(df["gaze_projected_to_right_view_x"], df["gaze_projected_to_right_view_y"], bins=10)
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

plt.clf()
plt.imshow(heatmap.T, extent=extent, origin='lower')
plt.show()
```
